```python
x=9
r=x%2
if r==0:
    print("Even number")
    
if r!=0:
    print('Odd number')
```

```python
#if-else
x=8
r=x%2
if r==0:
    print("Even number")
    
else:
    print('Odd number')
```

```python
#nested if
x=3
r=x%2
if r==0:
    print('Even number')
    if x>5:
        print('greater number')
else:
    print('Odd number')
```

```python
x=4
r=x%2
if r==0:
    print('Even number')
    if x>5:
        print('greater number')
    else:
        print('lesser number')
else:
    print('Odd number')
```
